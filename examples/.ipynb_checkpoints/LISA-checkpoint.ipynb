```julia
using GWSC
lisa = LISA();
lisa1 = LISA(transfer_file="RLISA1");
```

## Figure 2 

```julia
f = 10 .^ (-5:0.001:0)
R = lisa.R.(f)
R1 = lisa1.R.(f)

loglog(f, R)
loglog(f, R1)

legend(["numerical", "fit"])
xlabel("Frequency (Hz)")
ylabel("R ")
show()
```
