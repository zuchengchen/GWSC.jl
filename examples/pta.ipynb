## General Plots

```julia
pta1 = PTA(t=YEAR/20);
@time plotPI(pta1, ploteff=true, plotPILines=true)
```

```julia
pta2 = PTA(t=14*DAY, RMS=1e2, NP=36, TObs=1.);
@time plotPI(pta2, ploteff=true, plotPILines=true,
    PlotRange=(1e-9, 1e-4))
```
